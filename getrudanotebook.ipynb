```python
#empirical cumulative distribution function
def ecdf(data):
    
    n = len(data)
    x = np.sort(data)  #x-data for ecdf
    y = np.arange(1,n+1)/n
    
    return x,y
```

```python
# generate x and y data for ecdf
x_flips, y_flips = ecdf(x)

#plot ecdf
plt.plot(x_flips,y_flips,marker = '*',linestyle = 'none');
```
